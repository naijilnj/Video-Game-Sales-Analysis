### Distribution of Video Game Genres

We will examine the frequency or count of each genre in the dataset to understand the popularity and prevalence of different video game genres. This analysis will provide insights into the preferences of gamers and the types of games that are commonly developed and played.

```python
ax = sns.countplot( x='Genre', data = df)
sns.set(rc={'figure.figsize':(50,10)})
sns.set(font_scale=3)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
genre_counts = df['Genre'].value_counts()
fig, ax = plt.subplots(figsize=(20, 20))
pie = ax.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=90)

for text in pie[1]:
    text.set_fontsize(12)

ax.axis('equal')

plt.rcParams['figure.figsize'] = plt.rcParamsDefault['figure.figsize']
```

The data from the graphs and pie chart convincingly establish that the action genre holds a prominent position in the gaming landscape.

## SALES

### Yearly Global Sales

We will examine the total sales figures for video games on a yearly basis to understand the overall sales trends and patterns over time.

By analyzing the yearly global sales, we can gain insights into the growth of the video game industry, identify peak sales years, and detect any significant fluctuations or changes in sales patterns.

## Comparison of Sales Across Different Regions

We will examine the sales figures for North America (NA), Europe (EU), Japan (JP), and other regions over a specific period of time.
We will create a line chart to visualize the sales figures for each region over time. The x-axis of the line chart will represent the years or time periods, while the y-axis will indicate the sales values. Each region (NA, EU, JP, Other) will be represented by a separate line on the chart.

```python
sns.lineplot(x='Year', y='NA_Sales', data=NA_Yearly_Sales, label ='NA Sales')
sns.lineplot(x='Year', y='EU_Sales', data=EU_Yearly_Sales, label ='EU Sales')
sns.lineplot(x='Year', y='JP_Sales', data=JP_Yearly_Sales, label ='JP Sales')
sns.lineplot(x='Year', y='Other_Sales', data=Others_Yearly_Sales, label ='Other Regions Sales')
sns.set(rc={'figure.figsize':(10,5)})
sns.set(font_scale=2)
plt.legend
plt.gca().set_ylabel('Sales')
plt.show()
```

North America's highest sales occurred in 2008, Europe's highest sales occurred in 2009, Japan's highest sales occurred in 2006 and Other Region's highest Sales occurred in 2008.

The data underscores the widespread appeal of action games in North America, Europe and other regions, while emphasizing the strong influence of RPGs in the Japanese market.

Sales in North America performed significantly better than Europe, Japan and other Regions, Especially between 2000-2015

#### Distribution of Global Sales by Region

The analysis we are going to perform using a pie chart is focused on visualizing the distribution of global video game sales across different regions.

```python
TGS = df['Global_Sales'].sum()
TNAS = df['NA_Sales'].sum()
TEUS = df['EU_Sales'].sum()
TJPS = df['JP_Sales'].sum()
TOS = df['Other_Sales'].sum()

sales = [TNAS/TGS, TEUS/TGS, TJPS/TGS, TOS/TGS]
labels = ['North America', 'Europe', 'Japan', 'Other Regions']

fig, ax = plt.subplots(figsize=(10, 10))
pie = ax.pie(sales, labels=labels, autopct='%1.1f%%', startangle=90)

for text in pie[1]:
    text.set_fontsize(12)

ax.axis('equal')
plt.show()
```
